```python
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
auto_ts = YouTubeTranscriptApi.list_transcripts("lsWndfzuOc4").find_generated_transcript(['en'])
```

```python
time_spkr_tps = [('0:15:10','JBP'),('0:46:14','SVZ'),('1:20:51', 'JBP'),('1:33:00','SVZ'),
                ('1:46:30','END')]
```

### Now, we sub the base time to get timepoint:

```python
import datetime as dt
```

```python
base_time = dt.datetime(1900, 1, 1)
```

# Loop and build transcripts
We essentially loop over each speaker's segment & get what they said chunked out

```python
auto_ts_corp = auto_ts.fetch()
```

```python
# init dict
corp_dict = {'JBP':[],'SVZ':[]}
```

```python
curr_tup = time_spkr_tps.pop(0)
time_start = (dt.datetime.strptime(curr_tup[0], "%H:%M:%S") - base_time).total_seconds()
speaker = curr_tup[1]
auto_ts_corp = [x for x in auto_ts_corp if x['start'] > time_start]
```

```python
curr_tup
```

```python
for tup in time_spkr_tps:
    time_end = (dt.datetime.strptime(tup[0], "%H:%M:%S") - base_time).total_seconds()
    # get list of vals
    corp_dict[speaker].extend([x['text'] for x in auto_ts_corp if \
                               (x['start'] >= time_start) and (x['start'] < time_end)])
    
    time_start = time_end
    speaker = tup[1]
    
    
```

```python
corp_dict
```

# Bag the words
Now we combine Zizek & JBP's sections:

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
full_corpus = [' '.join(corp_dict['SVZ']), ' '.join(corp_dict['JBP'])]
```
